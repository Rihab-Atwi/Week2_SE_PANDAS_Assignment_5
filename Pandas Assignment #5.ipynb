```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.head(10)
```

```python
#custom function
def price_per_sqft(row):
    return row['price'] / row['sqft_living']
```

```python
# Apply the function using lambda
df['price_per_sqft'] = df.apply(lambda row: price_per_sqft(row), axis=1)

df.head()
```

```python
#function to categorize houses based on price
def price_range(price):
    if price <= 250000:
        return 'Low'
    elif price <= 500000:
        return 'Medium'
    else:
        return 'High'
```

```python
# Apply thefunction using lambda
df['price_range'] = df['price'].apply(lambda price: price_range(price))

df.head()
```

```python
#function to categorize size
def size(sqft):
    if sqft <= 1500:
        return 'Small'
    elif sqft <= 2500:
        return 'Medium'
    else:
        return 'Large'
```

```python
# Apply the function using lambda 
df['size_category'] = df['sqft_living'].apply(lambda sqft: size(sqft))
df.head()
```

```python
current_year = 2023
```

```python
#lambda function to calculate the age of a house
df['age'] = df['yr_built'].apply(lambda year_built: current_year - year_built)
df.head()
```
